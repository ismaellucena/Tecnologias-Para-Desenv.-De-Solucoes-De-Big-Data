```
import pandas as pd
import numpy as np


df = pd.read_csv('dados.csv', sep=';', engine='python')
```

```

print("Informações Iniciais:")
df.info()
print(df.head())
```

```

df_copia = df.copy()
```

```
df_copia['Calories'] = df_copia['Calories'].fillna(0)
```

```
df_copia['Date'] = df_copia['Date'].fillna('1900/01/01')
df_copia['Date'] = df_copia['Date'].replace('1900/01/01', np.nan)
```

```
df_copia['Date'] = df_copia['Date'].str.replace("'", "")
df_copia['Date'] = df_copia['Date'].replace('20201226', '2020/12/26')
df_copia['Date'] = pd.to_datetime(df_copia['Date'])
```

```

df_copia.dropna(subset=['Date'], inplace=True)
```

```
print("\nDataFrame Tratado com Sucesso:")
print(df_copia.to_string())
```
